```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import imageio
from IPython.display import HTML
```

```python
# we create list of countries have population more than 10000000 ppl in 2015
def list1_countries(pop):
    pop = pop[pop[2015] >= 10_000_000]
    return list(pop['country'])
```

Read Data

```python
df_cont = pd.read_csv('data/continents.csv', delimiter = ';')
```

```python
df_fert = pd.read_csv('data/fertility.csv', index_col = 0)
df_fert.index.name = 'country'
df_fert.dropna(inplace=True)
df_fert.columns = df_fert.columns.astype(int)
df_fert.shape
```

```python
df_fert
```

```python
df_lexp = pd.read_csv('data/lifeexpectancy.csv', index_col = 0)
df_lexp.index.name = 'country'
df_lexp.dropna(inplace=True)
# delete 2016 year
del df_lexp['2016']
df_lexp.shape
```

```python
df_lexp
```

```python
df_pop = pd.read_csv('data/population.csv', delimiter = ',', index_col = 0)
df_pop.index.name = 'country'
df_pop.columns = df_pop.columns.astype(int)
```

```python
# merge with continents
df_pop = pd.merge(df_cont, df_pop, how = 'inner', on = 'country')
df_pop.dropna(inplace=True)
df_pop.head()
```

```python
# calculate max ferility rate and max life exp to fix x&y coordinate
lma = int(df_lexp.max().max())+5
fma = int(df_fert.max().max()+1)  
```

```python
# make a set of countries with population more then 1000000 in 2015 to cut small countries
# and range of years
countries = list1_countries(df_pop)
years = list(df_pop.columns)
del years[0:2]
```

```python
df_fert = df_fert.reset_index()
df_lexp = df_lexp.reset_index()
```

```python
# melting dataframe
fert = df_fert.melt(id_vars='country', var_name='year', value_name='fertility_rate')
lexp = df_lexp.melt(id_vars='country', var_name='year', value_name='life_expectancy')
pop = df_pop.melt(id_vars=['continent', 'country'], var_name='year', value_name='population')
lexp['year'] = pd.to_numeric(lexp['year'])
```

```python
fert
```

```python
pop
```

```python
lexp
```

```python
df = pd.merge(pop, lexp, on = ['country','year'])
```

```python
df = pd.merge(df, fert, on = ['country','year'])
```

```python
df
```

```python
df = df.dropna()
```

```python
df
```

```python
# set up style for seaborn
sns.set(style="whitegrid", rc={'figure.figsize':(10,8)})
```

```python
# Create pmg pictures from our plots
df_subset = df.loc[df['country'].isin(countries)]
df_subset = df_subset.loc[df_subset['year'].isin(years)]
for i in years:
#     plt.figure(figsize=(10,5))
    plt.xlim(0, fma+3)
    plt.ylim(0, lma)
    plt.title('Year {}'.format(i))    
    plot = sns.scatterplot(
        data=df_subset.loc[df['year']==i], x="fertility_rate", y="life_expectancy", 
        hue="continent", size="population",
        sizes=(60, 300), alpha=0.8)
    plt.legend(bbox_to_anchor=(0.725, 1), loc=2, borderaxespad=0.)
    fig = plot.get_figure()
    fig.savefig('pics/output{}.png'.format(i)) 
    plt.clf()
```

```python
# Create Animated plot gif file
images = []

for i in years:
    filename = 'pics/output{}.png'.format(i)
    images.append(imageio.imread(filename))

imageio.mimsave('output.gif', images, fps=5)
```
